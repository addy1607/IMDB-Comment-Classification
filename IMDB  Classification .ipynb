```python
import pandas as pd

messages = pd.read_csv("imdb_categories_1.csv")
```

```python
messages
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['comment'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
# Creating the Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500)
X = cv.fit_transform(corpus).toarray()
```

```python
#Data preprocessing
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
messages["CategoryName"]=label_encoder.fit_transform(messages["CategoryName"])
```

```python
messages.head()
```

```python
messages["CategoryName"].unique()
```

```python
y=messages["CategoryName"]
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.naive_bayes import MultinomialNB
detect_model = MultinomialNB().fit(X_train, y_train)
```

```python
#prediction
y_pred=detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred,y_test))
```

```python
# Creating the TFIDF model
from sklearn.feature_extraction.text import TfidfVectorizer
tv = TfidfVectorizer(max_features=2500)
X = tv.fit_transform(corpus).toarray()
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.naive_bayes import MultinomialNB
detect_model = MultinomialNB().fit(X_train, y_train)
```

```python
#prediction
y_pred=detect_model.predict(X_test)
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred,y_test))
```
